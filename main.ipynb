```
# create a list of categorical columns to iterate over in our case it will be cbwd1
cat_cols = ['cbwd1']

def encode(data):
    '''function to encode non-null data and replace it in the original data'''
    #retains only non-null values
    nonulls = np.array(data.dropna())
    #reshapes the data for encoding
    impute_reshape = nonulls.reshape(-1,1)
    #encode date
    impute_ordinal = encoder.fit_transform(impute_reshape)
    #Assign back encoded values to non-null values
    data.loc[data.notnull()] = np.squeeze(impute_ordinal)
    return data

#create a for loop to iterate through each column in the data
for columns in cat_cols:
    encode(dane[columns])
```

Budowa i porównanie modeli wymagających feature scaling

```
# loading neccesary transformers

transformer_no_scaler = Pipeline(
   steps=
   [
    ('imputer', SimpleImputer(strategy='mean')),
    # ('scaler', MinMaxScaler())
   ]
)
cat_tran = Pipeline(
   steps=
   [
    # ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encode', OneHotEncoder())
   ]
)

from sklearn.compose import ColumnTransformer

columns_to_transform = ['year', 'month', 'day', 'year', 'hour', 'DEWP', 'TEMP',
                        'PRES', 'Iws', 'Is', 'Ir', 'pm2.5_lag_1h', 'week_day']

cat_column_to_transform = ['cbwd1']


preprocess_no_scaler = ColumnTransformer(
   transformers=[
                ('numeric', transformer_no_scaler, columns_to_transform),
                ('categorical', cat_tran, cat_column_to_transform)
  ]
)
# X_prepared = preprocess_no_scaler.fit_transform(X)
```

```
# modele niewymagające  feature scaling
classifiers_for_non_scaling = [
    ("RF", RandomForestRegressor(max_depth=15, n_jobs = -1, n_estimators=300, random_state=42)),
    ("LGBM", lightgbm.LGBMRegressor(learning_rate=0.07, n_estimators=1000)),
    ("DT", DecisionTreeRegressor()),
    ("XGBoost", XGBRegressor(learning_rate=0.01, n_estimators=1000)),
    ("CatBoost", CatBoostRegressor(iterations=500, learning_rate=0.1, max_bin=5000, max_depth=10, silent=True))
    ]

for classifier in classifiers_for_non_scaling:
    pipe = Pipeline(steps=[
                            ('preprocessor', preprocess_no_scaler),
                            ('classifier', classifier[1])
                          ])
    # pipe.fit(X_train, y_train)   
    # print("Test model score: %.3f" % pipe.score(X_test, y_test))
    scores_r2 = cross_val_score(pipe, X, y, scoring='r2', cv=5)
    # print(scores_r2)
    # print(f'Średnie dopasowanie modelu {classifier[0]} dla k = 5 wynosi: ', statistics.mean(scores_r2))
    scores_mae = cross_val_score(pipe, X, y, scoring='neg_mean_squared_error')
    print(f"Model {classifier[0]}, RMSE: ", sqrt(-statistics.mean(scores_mae)))
```

```
# Budowa modelu poddawanego analizie
catboost = CatBoostRegressor(iterations=500,learning_rate=0.07, max_bin=5000, max_depth=8, silent=True)
catboost.fit(X_train,y_train)
```

```
import shap
# Obliczanie wartości Shapa dla całego zbioru treningowego
shap_values = shap.TreeExplainer(catboost).shap_values(X_train)
shap.summary_plot(shap_values, X_train, plot_type="bar")
```
