```
# instantiate both packages to use
encoder = OrdinalEncoder()
# instatiate the knn imputer
imputer = KNN()
```

```
# create a list of categorical columns to iterate over in our case it will be cbwd1
cat_cols = ['cbwd1']

def encode(data):
    '''function to encode non-null data and replace it in the original data'''
    #retains only non-null values
    nonulls = np.array(data.dropna())
    #reshapes the data for encoding
    impute_reshape = nonulls.reshape(-1,1)
    #encode date
    impute_ordinal = encoder.fit_transform(impute_reshape)
    #Assign back encoded values to non-null values
    data.loc[data.notnull()] = np.squeeze(impute_ordinal)
    return data

#create a for loop to iterate through each column in the data
for columns in cat_cols:
    encode(dane[columns])
```

```
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
```

```
# loading neccesary transformers

transformer = Pipeline(
   steps=
   [
    # ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', MinMaxScaler())
   ]
)
cat_tran = Pipeline(
   steps=
   [
    # ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encode', OneHotEncoder())
   ]
)
```

```
from sklearn.compose import ColumnTransformer

columns_to_transform = ['year', 'month', 'day', 'year', 'hour', 'DEWP', 'TEMP',
                        'PRES', 'Iws', 'Is', 'Ir', 'pm2.5_lag_1h', 'week_day']

cat_column_to_transform = ['cbwd1']


preprocess = ColumnTransformer(
   transformers=[
                ('numeric', transformer, columns_to_transform),
                ('categorical', cat_tran, cat_column_to_transform)
  ]
)
# X_prepared = preprocess.fit_transform(X)
```

```
# loading neccesary transformers

transformer_no_scaler = Pipeline(
   steps=
   [
    ('imputer', SimpleImputer(strategy='mean')),
    # ('scaler', MinMaxScaler())
   ]
)
cat_tran = Pipeline(
   steps=
   [
    # ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encode', OneHotEncoder())
   ]
)

from sklearn.compose import ColumnTransformer

columns_to_transform = ['year', 'month', 'day', 'year', 'hour', 'DEWP', 'TEMP',
                        'PRES', 'Iws', 'Is', 'Ir', 'pm2.5_lag_1h', 'week_day']

cat_column_to_transform = ['cbwd1']


preprocess_no_scaler = ColumnTransformer(
   transformers=[
                ('numeric', transformer_no_scaler, columns_to_transform),
                ('categorical', cat_tran, cat_column_to_transform)
  ]
)
# X_prepared = preprocess_no_scaler.fit_transform(X)
```
